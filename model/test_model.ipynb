```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
import random
import os
```

```python
model = tf.keras.models.load_model('./Saved_model/VeryVeryBest_ResNet50_20231127_204650/H5Model_ResNet50_LastEpoch.h5')
```

```python
def prediction_picture(image_path):
    img = image.load_img(image_path)
    img_arr = image.img_to_array(img)
    img_for_prediction = np.expand_dims(img_arr, axis=0)

    prediction = model.predict(img_for_prediction)
    predicted_labels = np.argmax(prediction, axis=1)

    class_labels = os.listdir("./dataset/original/")

    print(f"prediction accuracy : {prediction[0][predicted_labels][0]*100:.2f}% {class_labels[predicted_labels[0]]}")
    plt.imshow(img)
    plt.show()
```

```python
DATA_DIR = './dataset/test/'
random_class = random.choice(os.listdir(DATA_DIR))
random_img = random.choice(os.listdir(DATA_DIR + random_class))
img_path = os.path.join(DATA_DIR, random_class, random_img)
print("Real is", random_class)
prediction_picture(img_path)
```
